## filling different values in null in different columns

```python
import pandas as pd
df=pd.read_csv("employees.csv")
df
```

```python
df7=df.fillna({'Team' : 'abc', 
              'Gender':'def'})
df7
```

```python
df8=df.fillna(method= 'pad', axis = 1)
df8
```

# Filling the null values with the mean of a column


```python
df10=df.fillna(value=df['Salary'].mean())
df10
```

```python
df11=df.fillna(value=df['Salary'].max())
df11
```

```python
df12=df.fillna(value=df['Salary'].min())
df12
```
