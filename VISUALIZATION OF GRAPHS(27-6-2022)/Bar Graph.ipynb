# Bar Graph

```python
import matplotlib.pyplot as plt
import numpy as np
country=["USA","INDIA","CHINA","PARIS","NEWYORK"]
Gdp_per_capita=[4800,38000,5000,45000,44000]
plt.bar(country,Gdp_per_capita)
plt.title("COUNTRY VS GDP PER CAPITA")
plt.xlabel('country')
plt.ylabel('Gdp per capita')
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np
country=["Thailand","Germany","Georgia","Uk","France"]
Gdp_per_capita=[4800,38000,5000,45000,44000]
plt.barh(country,Gdp_per_capita)
plt.title("COUNTRY VS GDP PER CAPITA")
plt.xlabel("COUNTRY")
plt.ylabel("GDP PER CAPITA")
plt.show()
```
