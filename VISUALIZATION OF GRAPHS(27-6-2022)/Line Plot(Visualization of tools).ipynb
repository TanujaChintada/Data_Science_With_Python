```python
import matplotlib.pyplot as plt
import numpy as np
plt.title("Line plot")
plt.xlabel("xaxis")
plt.ylabel("yaxis")
xpoints=np.array([0,20])
ypoints=np.array([0,50])
plt.plot(xpoints,ypoints)
plt.show()
```

```python
plt.plot([5,100],[5,200],"--")
plt.show()
```

```python
plt.plot([5,90],[8,200],"--")
plt.show()
```

```python
x=[15,25,35,46,58]
y=[10,20,30,40,50]
plt.scatter(x,y)
plt.show()
```

## multiple points

```python
x=np.array([20,40,60])
y=np.array([68,98,34])
plt.plot(x,y)
plt.show()
```

```python
xpoints=np.array([1,2,6,8])
ypoints=np.array([3,8,1,10])
plt.plot(xpoints,ypoints)
plt.show()
```

## plotting without xpoints

```python
ypoints=np.array([3,8,1,10,5,7])
plt.plot(ypoints)
plt.show()
```

## plotting without xpoints

```python
ypoints=np.array([3,8,1,10,5,7])
plt.plot(ypoints,marker="o",mec="r")
plt.show()
```

```python
ypoints=np.array([30,45,67,98])
plt.plot(ypoints,marker="*",mec="b")
plt.show()
```

## Color

```python
ypoints=np.array([3,8,1,10])
plt.plot(ypoints,marker="*",mec="g",color="r")
plt.show()
```

```python
ypoints=np.array([3,8,1,10])
plt.plot(ypoints,marker="o",mec="b",color="b",linewidth="15")
plt.show()
```
