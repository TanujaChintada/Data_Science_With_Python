# pie chart

```python
import matplotlib.pyplot as plt
import numpy as np
y=np.array([35,50,15])
plt.pie(y)
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np
x=np.array([35,25,25,15])
col=["r","b","pink","r"]
plt.pie(x,colors=col,labels=["CD","WT","DSP","COA"])
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np
y=np.array([35,25,15,25])
myexplode=[0.3,0.4,0.5,0.2]
plt.pie(y,labels=["CD","COA","DSP","WT"],explode=myexplode)
plt.legend()
plt.legend(title="subjects")
plt.show()
```
