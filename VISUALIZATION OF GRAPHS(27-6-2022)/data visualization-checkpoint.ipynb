```python
import pandas as pd
import matplotlib.pyplot as plt
df=pd.read_csv("company_sales_data.csv")
df
```

```python
profitlist = df["total_profit"].tolist()
monthlist = df["month_number"].tolist()
plt.plot(monthlist,profitlist,label="monthwise profit data of last year")
plt.xlabel("monthnumber")
plt.ylabel("total profit")
plt.title("monthnumber vs profitlist")
plt.xticks(monthlist)
plt.yticks([100000,200000,300000,400000])
plt.legend(loc="lower right")
plt.show()
```

```python
monthlist = df["month_number"].tolist()
profitlist = df["total_profit"].tolist()
plt.plot(monthlist,profitlist,label ="total profit of all months",color="r",marker="o",markerfacecolor="g",linestyle="--",linewidth=5)
plt.xlabel("monthnumber")
plt.ylabel("profitlist")
plt.legend(loc="upper left")
plt.title("company sales of the last year")
plt.show()
```

# Multiline Plot

```python
monthlist = df["month_number"].tolist()
fcsalesdata = df["facecream"].tolist()
fcwashdata = df["facewash"].tolist()
tpdata = df["toothpaste"].tolist()
shampoodata = df["shampoo"].tolist()
moisturizerdata = df["moisturizer"].tolist()
plt.plot(monthlist,fcsalesdata,label='fcsalesdata',marker="*",linewidth=3)
plt.plot(monthlist,fcwashdata,label='fcwashdata',marker="o",linewidth=4)
plt.plot(monthlist,tpdata,label='tpdata',marker=">",markerfacecolor="g",linewidth=5)
plt.plot(monthlist,shampoodata,label="shapmpoodata",marker="<",linewidth=6)
plt.plot(monthlist,moisturizerdata,label="moisturizerdata",marker="o",linewidth=4)
plt.xticks(monthlist)
plt.xlabel("monthnumber")
plt.ylabel("sales units in number")
plt.legend(loc="upper left")
plt.yticks([1000,2000,4000,6000,8000,10000,12000,15000,18000])
plt.show()
```

```python
monthlist = df["month_number"].tolist()           #grid plot
toothpastesalesdata = df["toothpaste"].tolist()
plt.scatter(monthlist,toothpastesalesdata)
plt.xlabel("monthlist")
plt.ylabel("toothpaste")
plt.title("tooth paste sales data")
plt.grid(True,linewidth=1,linestyle="--")
plt.show()
```

# bar graph

```python
monthlist = df["month_number"].tolist()
bathingsoapdata = df["bathingsoap"].tolist()
plt.bar(monthlist,bathingsoapdata)
plt.grid(True,linewidth=1,linestyle="--")
plt.show()
```

# pie chart

```python
monthlist = df["month_number"].tolist()
labels = ['facewash','facecream','toothpaste','bathingsoap','shampoo','moisturizer']
salesdata = [df['facewash'].sum(),df['facecream'].sum(),df['toothpaste'].sum(),df['bathingsoap'].sum(),df['shampoo'].sum(),df['moisturizer'].sum()]
plt.pie(salesdata,labels=labels,autopct='%1.1f%%')
plt.axis("equal")
plt.legend(loc="lower left")
plt.show()
```
